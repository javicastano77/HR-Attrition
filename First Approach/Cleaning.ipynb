```python
import pandas as pd

# Supongamos que ya tienes un DataFrame llamado 'df' con las columnas 'age', 'education', etc.

# Crear la variable de intervalos de edad
intervalos = [18, 30, 40, 50, 60, float('inf')]
etiquetas = ['18-29', '30-39', '40-49', '50-59', '60+']
df['age_interval'] = pd.cut(df['age'], bins=intervalos, labels=etiquetas, right=False)

# Mapear variables categ√≥ricas
mapp_2 = {1: 'Below College', 2: 'College', 3: 'Bachelor', 4: 'Master', 5: 'Doctor'}
df['education_cat'] = df['education'].map(mapp_2)

mapp_3 = {1: 'Low', 2: 'Medium', 3: 'High', 4: 'Very High'}
df['environment_satisf_cat'] = df['environment_satisf'].map(mapp_3)

mapp_4 = {1: 'Low', 2: 'Medium', 3: 'High', 4: 'Very High'}
df['job_involvement_cat'] = df['job_involvement'].map(mapp_4)

mapp_5 = {1: 'associate', 2: 'analyst', 3: 'specialist', 4: 'manager', 5: 'director'}
df['job_level'] = df['job_level'].map(mapp_5)

mapp_6 = {1: 'Low', 2: 'Medium', 3: 'High', 4: 'Very High'}
df['job_satisf_cat'] = df['job_satisfaction'].map(mapp_6)

mapp_7 = {1: 'Bad', 2: 'Good', 3: 'Better', 4: 'Best'}
df['worklife_balance_cat'] = df['worklife_balance'].map(mapp_7)

mapp_8 = {1: 'Low', 2: 'Good', 3: 'Excellent', 4: 'Outstanding'}
df['performance_rating_cat'] = df['performance_rating'].map(mapp_8)

mapp_9 = {1: 'Low', 2: 'Medium', 3: 'High', 4: 'Very High'}
df['relationship_satisf_cat'] = df['relationship_satisf'].map(mapp_9)

df
```
